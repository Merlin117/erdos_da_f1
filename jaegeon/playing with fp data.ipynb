```python
# Merge the two DataFrames
# Merging on raceId, year, round, driverNumber, and driverId
# Using suffixes to distinguish 'positionNumber' from each DataFrame
merged_df = pd.merge(
    race_results[['raceId', 'year', 'round', 'driverNumber', 'driverId', 'positionNumber']],
    starting_grid[['raceId', 'year', 'round', 'driverNumber', 'driverId', 'positionNumber']],
    on=['raceId', 'year', 'round', 'driverNumber', 'driverId'],
    how='left',
    suffixes=('_race_results', '_starting_grid')
)

# Calculate 'positionDifference'
# If either 'positionNumber_race_results' or 'positionNumber_starting_grid' is NaN,
# the result of the subtraction will automatically be NaN.
merged_df['positionDifference'] = merged_df['positionNumber_race_results'] - merged_df['positionNumber_starting_grid']

# Select the desired columns for the final DataFrame
final_df = merged_df[['raceId', 'year', 'round', 'driverId', 'positionDifference']]

# Save the final DataFrame to a CSV file
final_df.to_csv('race-positions-difference.csv', index=False)
```

```python
qualifying = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\data_f1db\\f1db-races-qualifying-results.csv')
```

```python
# Prepare qualifying data: select relevant columns and rename position column
qualifying_positions = qualifying[['raceId', 'year', 'round', 'driverId', 'positionNumber']].copy()
qualifying_positions.rename(columns={'positionNumber': 'qualifying_position'}, inplace=True)

# Dictionary of free practice dataframes and their names for easy iteration
fp_dfs = {
    'Free Practice 1': fp1,
    'Free Practice 2': fp2,
    'Free Practice 3': fp3
}

# Iterate through each free practice DataFrame, merge, calculate difference, and plot
for fp_name, fp_df in fp_dfs.items():
    # Prepare free practice data: select relevant columns and rename position column
    fp_positions = fp_df[['raceId', 'year', 'round', 'driverId', 'positionNumber']].copy()
    fp_positions.rename(columns={'positionNumber': 'fp_position'}, inplace=True)

    # Merge free practice and qualifying data
    # Inner join ensures we only compare for races/drivers present in both
    merged_df = pd.merge(
        fp_positions,
        qualifying_positions,
        on=['raceId', 'year', 'round', 'driverId'],
        how='inner'
    )

    # Calculate the new difference: qualifying position - free practice position
    merged_df['qualifying_fp_difference'] = merged_df['qualifying_position'] - merged_df['fp_position']

    # Drop rows where the difference is NaN (if either original position was NaN)
    merged_df.dropna(subset=['qualifying_fp_difference'], inplace=True)

    # Convert year to category for plotting, though seaborn handles int years well for boxplots
    merged_df['year'] = merged_df['year'].astype(str)

    # Create the box and whisker plot
    plt.figure(figsize=(15, 7))
    sns.boxplot(
        data=merged_df,
        x='year',
        y='qualifying_fp_difference',
        order=sorted(merged_df['year'].unique()) # Ensure years are in order
    )
    plt.title(f'Distribution of Qualifying Position - {fp_name} Position by Year')
    plt.xlabel('Year')
    plt.ylabel('Qualifying Position - Free Practice Position')
    plt.xticks(rotation=90) # Rotate x-axis labels for better readability
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()
```

```python
race_position_diff_df = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\jaegeon\gridposition-result-difference.csv')

# Group by raceId, year, round and calculate the mean of 'positionDifference'
avg_pos_change_per_race_df = race_position_diff_df.groupby(['raceId', 'year', 'round'])['positionDifference'].mean().reset_index()
avg_pos_change_per_race_df.rename(columns={'positionDifference': 'avg_pos_change'}, inplace=True)

# List of free practice dataframes and their names for easy iteration
fp_dfs = {
    'Free Practice 1': fp1,
    'Free Practice 2': fp2,
    'Free Practice 3': fp3
}

# C. Merge and Plot
for fp_name, fp_df in fp_dfs.items():
    # Calculate Fastest and Slowest Time per Race and the Spread
    fp_race_times = fp_df.groupby(['raceId', 'year', 'round'])['timeMillis'].agg(
        min_timeMillis='min',
        max_timeMillis='max'
    ).reset_index()

    # Calculate spread: (fastest time / slowest time * 100)
    fp_race_times['fp_spread'] = (fp_race_times['min_timeMillis'] / fp_race_times['max_timeMillis']) * 100

    # Merge with the average position change data
    merged_plot_df = pd.merge(
        fp_race_times[['raceId', 'year', 'round', 'fp_spread']],
        avg_pos_change_per_race_df,
        on=['raceId', 'year', 'round'],
        how='inner' # Only include races where both data types are available
    )

    # Create the scatter plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        data=merged_plot_df,
        x='fp_spread',
        y='avg_pos_change',
        alpha=0.7
    )
    plt.title(f'Average Spread of {fp_name} Times vs. Average Race Position Change per Race')
    plt.xlabel(f'Average {fp_name} Time Spread (Fastest / Slowest * 100)')
    plt.ylabel('Average Race Position Change (Race Pos - Grid Pos)')
    plt.grid(True)
    plt.axhline(0, color='grey', linestyle='--', linewidth=0.8) # Add a line at y=0 for reference
    plt.show()
```

```python
# A. Prepare the Y-axis data (Average Change in Position during Race per Race and Circuit)
# 1. Load race-positions-difference.csv
race_position_diff_df = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\jaegeon\gridposition-result-difference.csv')

# 2. Load f1db-races.csv
races_df = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\data_f1db\\f1db-races.csv')

# *** CORRECTED STEP: Rename 'id' column in races_df to 'raceId' for merging ***
races_df.rename(columns={'id': 'raceId'}, inplace=True)

# Select relevant columns from races_df that include the (now renamed) 'raceId'
race_circuit_info = races_df[['raceId', 'year', 'round', 'circuitId']].copy()

# 3. Merge race-positions-difference_df with circuit info to get circuitId
merged_race_pos_diff_circuit = pd.merge(
    race_position_diff_df,
    race_circuit_info,
    on=['raceId', 'year', 'round'],
    how='inner'
)

# 4. Calculate Average Change in Position per Race and Circuit
avg_pos_change_per_race_circuit_df = merged_race_pos_diff_circuit.groupby(['raceId', 'year', 'round', 'circuitId'])['positionDifference'].mean().reset_index()
avg_pos_change_per_race_circuit_df.rename(columns={'positionDifference': 'avg_pos_change'}, inplace=True)

# B. Prepare the X-axis data (Average Spread of Free Practice Results per Race and Circuit)
# List of free practice dataframes and their names for easy iteration
fp_dfs = {
    'Free Practice 1': fp1,
    'Free Practice 2': fp2,
    'Free Practice 3': fp3
}

# C. Merge and Plot
for fp_name, fp_df in fp_dfs.items():
    # Merge current FP df with races_df (which now has 'raceId') to get circuitId
    fp_df_with_circuit = pd.merge(
        fp_df,
        race_circuit_info, # Use the race_circuit_info derived from f1db-races.csv
        on=['raceId', 'year', 'round'],
        how='inner'
    )

    # Calculate Fastest and Slowest Time per Race and Circuit from current FP df with circuitId
    fp_race_times = fp_df_with_circuit.groupby(['raceId', 'year', 'round', 'circuitId'])['timeMillis'].agg(
        min_timeMillis='min',
        max_timeMillis='max'
    ).reset_index()

    # Calculate spread: (fastest time / slowest time * 100)
    fp_race_times['fp_spread'] = (fp_race_times['min_timeMillis'] / fp_race_times['max_timeMillis']) * 100

    # Merge FP spread data with average position change data (which also has circuitId)
    merged_plot_data_race_level = pd.merge(
        fp_race_times[['raceId', 'year', 'round', 'circuitId', 'fp_spread']],
        avg_pos_change_per_race_circuit_df,
        on=['raceId', 'year', 'round', 'circuitId'],
        how='inner'
    )

    # Calculate the ratio: (fp_spread / avg_pos_change)
    # Handle division by zero and inf values
    merged_plot_data_race_level['ratio'] = merged_plot_data_race_level['fp_spread'] / merged_plot_data_race_level['avg_pos_change']
    merged_plot_data_race_level.replace([np.inf, -np.inf], np.nan, inplace=True) # Replace inf with NaN
    merged_plot_data_race_level.dropna(subset=['ratio'], inplace=True) # Drop NaNs

    # Aggregate ratio by circuitId
    avg_ratio_per_circuit = merged_plot_data_race_level.groupby('circuitId')['ratio'].mean().reset_index()
    avg_ratio_per_circuit.sort_values(by='circuitId', inplace=True)

    # Create the bar plot
    plt.figure(figsize=(18, 8)) # Increased figure size for better readability of circuit IDs
    sns.barplot(
        data=avg_ratio_per_circuit,
        x='circuitId',
        y='ratio',
        palette='viridis'
    )
    plt.title(f'Average ({fp_name} Time Spread / Average Race Position Change) per Circuit', fontsize=16)
    plt.xlabel('Circuit ID', fontsize=14)
    plt.ylabel(f'Average ({fp_name} Spread / Race Pos Change)', fontsize=14)
    plt.xticks(rotation=90, fontsize=10) # Rotate x-axis labels for better readability
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()
```

```python
# List of free practice dataframes and their names for easy iteration
fp_dfs = {
    'Free Practice 1': fp1,
    'Free Practice 2': fp2,
    'Free Practice 3': fp3
}

# 3. Iterate through FP1, FP2, FP3 and plot
for fp_name, fp_df in fp_dfs.items():
    # Merge current FP df with races_df to get circuitId
    fp_df_with_circuit = pd.merge(
        fp_df,
        race_circuit_info,
        on=['raceId', 'year', 'round'],
        how='inner'
    )

    # Calculate Fastest and Slowest Time per Race and Circuit
    # Ensure timeMillis is numeric and handle potential NaNs
    fp_df_with_circuit['timeMillis'] = pd.to_numeric(fp_df_with_circuit['timeMillis'], errors='coerce')
    fp_race_times = fp_df_with_circuit.groupby(['raceId', 'year', 'round', 'circuitId'])['timeMillis'].agg(
        min_timeMillis='min',
        max_timeMillis='max'
    ).reset_index()

    # Calculate spread: (fastest time / slowest time * 100)
    # Handle potential division by zero or NaN results before calculation
    fp_race_times['fp_spread'] = np.where(
        fp_race_times['max_timeMillis'] != 0,
        (fp_race_times['min_timeMillis'] / fp_race_times['max_timeMillis']) ,
        np.nan
    )
    fp_race_times.dropna(subset=['fp_spread'], inplace=True) # Drop rows where spread couldn't be calculated

    # Aggregate Spread by Circuit
    avg_fp_spread_per_circuit = fp_race_times.groupby('circuitId')['fp_spread'].mean().reset_index()
    avg_fp_spread_per_circuit.sort_values(by='circuitId', inplace=True)

    # *** CORRECTED STEP: Sort avg_fp_spread_per_circuit by 'fp_spread' in descending order ***
    avg_fp_spread_per_circuit.sort_values(by='fp_spread', ascending=False, inplace=True)

    # Create the bar plot
    plt.figure(figsize=(18, 8)) # Increased figure size for better readability of circuit IDs
    sns.barplot(
        data=avg_fp_spread_per_circuit,
        x='circuitId',
        y='fp_spread',
        palette='magma' # Using a different palette
    )
    plt.title(f'Average {fp_name} Time Spread per Circuit', fontsize=16)
    plt.xlabel('Circuit ID', fontsize=14)
    plt.ylabel(f'Average {fp_name} Time Spread (Fastest / Slowest)', fontsize=14)
    plt.xticks(rotation=90, fontsize=10) # Rotate x-axis labels for better readability
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()
```

```python
# 1. Load Race Data (f1db-races.csv) and filter by year
races_df.rename(columns={'id': 'raceId'}, inplace=True)
races_df_1961_onwards = races_df[races_df['year'] >= 1961].copy()
race_circuit_info_1961_onwards = races_df_1961_onwards[['raceId', 'circuitId']].copy()

# 2. Load Race Results and Starting Grid
race_results_df = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\data_f1db\\f1db-races-race-results.csv')
starting_grid_df = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\data_f1db\\f1db-races-starting-grid-positions.csv')

# 3. Prepare Positions Data
race_results_subset = race_results_df[['raceId', 'driverId', 'positionNumber']].copy()
race_results_subset.rename(columns={'positionNumber': 'race_pos'}, inplace=True)

starting_grid_subset = starting_grid_df[['raceId', 'driverId', 'positionNumber']].copy()
starting_grid_subset.rename(columns={'positionNumber': 'grid_pos'}, inplace=True)

# 4. Count Starters per Race (from starting_grid_df)
num_starters_per_race = starting_grid_subset.groupby('raceId')['driverId'].nunique().reset_index(name='num_starters')

# 5. Merge Positions Data
merged_positions_df = pd.merge(
    race_results_subset,
    starting_grid_subset,
    on=['raceId', 'driverId'],
    how='inner'
)

# 6. Filter Merged Positions by Year (1961 onwards)
filtered_positions_df = pd.merge(
    merged_positions_df,
    races_df_1961_onwards[['raceId']],
    on='raceId',
    how='inner'
)

# Handle Missing Positions and Convert Positions to Integer
filtered_positions_df.dropna(subset=['race_pos', 'grid_pos'], inplace=True)
filtered_positions_df['race_pos'] = filtered_positions_df['race_pos'].astype(int)
filtered_positions_df['grid_pos'] = filtered_positions_df['grid_pos'].astype(int)

# 7. Calculate Overtakes per Race (for filtered races)
def count_overtakes_in_race(race_data):
    race_data_sorted_by_grid = race_data.sort_values(by='grid_pos').reset_index(drop=True)
    race_pos_order = race_data_sorted_by_grid['race_pos'].tolist()
    n = len(race_pos_order)
    overtake_count = 0
    for i in range(n):
        for j in range(i + 1, n):
            if race_pos_order[i] > race_pos_order[j]:
                overtake_count += 1
    return overtake_count

overtakes_per_race_1961_onwards = filtered_positions_df.groupby('raceId').apply(count_overtakes_in_race).reset_index(name='num_overtakes')

# 8. Merge Overtakes with Number of Starters (and filter by year)
# Ensure num_starters_per_race also only contains races from 1961 onwards
num_starters_per_race_1961_onwards = pd.merge(
    num_starters_per_race,
    races_df_1961_onwards[['raceId']],
    on='raceId',
    how='inner'
)

merged_overtakes_starters = pd.merge(
    overtakes_per_race_1961_onwards,
    num_starters_per_race_1961_onwards,
    on='raceId',
    how='inner'
)

# 9. Normalize Overtakes
merged_overtakes_starters['normalized_overtakes'] = np.where(
    merged_overtakes_starters['num_starters'] != 0,
    merged_overtakes_starters['num_overtakes'] / merged_overtakes_starters['num_starters'],
    np.nan
)
merged_overtakes_starters.dropna(subset=['normalized_overtakes'], inplace=True)

# 10. Count Races per Circuit (1961 onwards)
race_counts_per_circuit_1961_onwards = races_df_1961_onwards.groupby('circuitId')['raceId'].nunique().reset_index()
race_counts_per_circuit_1961_onwards.rename(columns={'raceId': 'race_count'}, inplace=True)

# 11. Filter Circuits (more than 10 races from 1961 onwards)
circuits_with_more_than_10_races_1961_onwards = race_counts_per_circuit_1961_onwards[
    race_counts_per_circuit_1961_onwards['race_count'] > 10
]['circuitId'].tolist()

# 12. Merge Normalized Overtakes with Circuit Info (1961 onwards)
normalized_overtakes_circuit_1961_onwards = pd.merge(
    merged_overtakes_starters,
    race_circuit_info_1961_onwards,
    on='raceId',
    how='inner'
)

# 13. Aggregate by Circuit (1961 onwards)
avg_normalized_overtakes_per_circuit = normalized_overtakes_circuit_1961_onwards.groupby('circuitId')['normalized_overtakes'].mean().reset_index()

# 14. Filter Aggregated Data (by 10+ races from 1961 onwards)
filtered_avg_normalized_overtakes_per_circuit = avg_normalized_overtakes_per_circuit[
    avg_normalized_overtakes_per_circuit['circuitId'].isin(circuits_with_more_than_10_races_1961_onwards)
].copy()

# Sort in descending order for plotting
filtered_avg_normalized_overtakes_per_circuit.sort_values(by='normalized_overtakes', ascending=False, inplace=True)

# 15. Create Bar Plot
plt.figure(figsize=(15, 8))
sns.barplot(
    data=filtered_avg_normalized_overtakes_per_circuit,
    x='circuitId',
    y='normalized_overtakes',
    palette='magma' # Using a different palette
)
plt.title('Average Normalized Overtakes per Circuit (Races On or After 1961, > 10 Races)', fontsize=16)
plt.xlabel('Circuit ID', fontsize=14)
plt.ylabel('Average Normalized Overtakes (Overtakes / Starters)', fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
